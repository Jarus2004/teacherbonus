```python
from sklearn.model_selection import train_test_split

x=df[['1st EXT. STEAM FLOW (Ton/hr)',
     '2nd EXT. STEAM FLOW (Ton/hr)',
     '3rd EXT. STEAM FLOW (Ton/hr)',
     'SURFACE CONDENSATE FLOW (Ton/hr)']]

y=df[['Utility (baht/Ton/hr)']]
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.neural_network import MLPRegressor
from sklearn.datasets import make_regression
from sklearn.model_selection import train_test_split
x,y=make_regression(n_samples=100,random_state=1)#ถ้าใช้random=1 ค่าx,yจะเท่ากันตลอดถึงแม้จะเปลียนmodel

#นำข้อมูลมาtrainและทดสอบmodel เพื่อให้เป็นค่าคงที่จึงเลือกใช้random=1
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state=1)
```

```python
mlp=MLPRegressor()
```

```python
mlp.fit(x_train,y_train)
```

```python
mlp.score(x_test,y_test)
```

```python
from sklearn.metrics import mean_absolute_error
y_pred= mlp.predict(x_test)

mae=mean_absolute_error(y_test,y_pred)
print(f"\nMAE: {mae:.2f}bath/ton/hr")
```

```python
from sklearn.metrics import mean_absolute_percentage_error
mape=mean_absolute_percentage_error(y_test,y_pred)
print(f"\nMAPE: {mape:.2f}%")
```
