```python
import pandas as pd
df= pd.read_csv("/content/drive/MyDrive/ML/linear/Data-for-study.csv",delimiter=",",encoding="utf-8")
df
```

```python
from sklearn.model_selection import train_test_split

x=df[['1st EXT. STEAM FLOW (Ton/hr)',
     '2nd EXT. STEAM FLOW (Ton/hr)',
     '3rd EXT. STEAM FLOW (Ton/hr)',
     'SURFACE CONDENSATE FLOW (Ton/hr)']]

y=df[['Utility (baht/Ton/hr)']]
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
```

```python
print('Traing Features Shape:',x_train.shape)
print('Traing LAbels Shape:',y_train.shape)
print('Traing Features Shape:',x_test.shape)
print('Traing LAbels Shape:',y_test.shape)
```


# **สร้างMODEL Randoms tree**

```python
from sklearn.ensemble import RandomForestRegressor
rf=RandomForestRegressor(n_estimators=1000,random_state=42)

#train the model on data
rf.fit(x_train,y_train)
```

# **ทดสอบความถูกต้อง**

# R-squared

```python
rf.score(x_test,y_test)
```

# **MAE**

```python
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_absolute_percentage_error
```

```python
rf.fit(x_train,y_train)

y_pred=rf.predict(x_test)

mae=mean_absolute_error(y_test,y_pred)
print(f"\nMAE: {mae:.2f}bath/ton/hr")
```

```python
mape=mean_absolute_percentage_error(y_test,y_pred)
print(f"\nMAPE: {mape:.2f}%")
```

# **ACCURACY**

```python
import numpy as np
maximum_y_true=np.max(df.iloc[:,17])
maximum_error=np.array(1*maximum_y_true)/100
y_pred_array=np.array(y_pred)
y_test_array=np.array(y_test)
y_pred_new=np.zeros((y_test_array.size,1))
dif = abs(y_test_array - y_pred_new)

i=0
for x in y_pred:
  y_pred_new[i]=x
  i+=1
```

```python
def accuracy_score(y_true,y_predict,percent):
  maximum_y_true= np.max(df.iloc[:,17])
  maximum_error = (percent*maximum_y_true)/100
  dif=np.array(abs(y_true-y_predict))
  n_total = y_true.size
  count =0
  for x in dif:
    if x <= maximum_error:
      count=1+count
  cal=count/n_total
  return cal
```

```python
accuracy_score(y_test_array,y_pred_new,5)
```
